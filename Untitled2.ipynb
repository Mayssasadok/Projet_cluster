La stemming est une sorte de normalisation des mots. La normalisation est une technique dans laquelle un ensemble de mots dans une phrase est converti en une sÃ©quence pour raccourcir sa recherche. ####  <a style="color:#FF0000" > Le clustering K-Means :  </a>
est un algorithme d'apprentissage automatique non supervisÃ©. Contrairement aux algorithmes traditionnels d'apprentissage automatique supervisÃ©, K-Means tente de classer les donnÃ©es sans avoir d'abord Ã©tÃ© formÃ© avec des donnÃ©es Ã©tiquetÃ©es. Une fois l'algorithme exÃ©cutÃ© et les groupes dÃ©finis, toute nouvelle donnÃ©e peut Ãªtre facilement affectÃ©e au groupe le plus pertinent. WCSS est dÃ©fini comme la somme de la distance au carrÃ© entre chaque membre du cluster et son centre de gravitÃ©

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(3,40):
    Kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0,verbose=True)
    Kmeans.fit(X)
    wcss.append(Kmeans.inertia_)
```

Il existe diffÃ©rentes mÃ©thodes pour dÃ©terminer le nombre optimal de grappes et l'une d'entre elles est la mÃ©thode du coude. L'approche consiste Ã  rechercher un coude dans le graphe WCSS. Habituellement, la partie du graphique avant le coude dÃ©croÃ®trait fortement, tandis que la partie aprÃ¨s serait plus lisse

```python

import matplotlib.pyplot as plt
plt.plot(range(3,40),wcss)
plt.xlabel('number of clusters')
plt.ylabel('word per cluster')
plt.show()
```

```python
true_k=40
Kmeans=KMeans(n_clusters=true_k,init='k-means++',n_init=1)
Kmeans.fit(X)
```

####  <a style="color:#FF0000" > Fonction de jaccard </a>
ona crÃ©er une base de donnÃ©es contenant le nombre total de tweets par catÃ©gorie et par personne. Une base de donnÃ©es 4D avec la colonne d'index remplie d'utilisateurs, et 3 autres colonnes contenant le nombre total de tweets de l'utilisateur dans les classes sociales, culturelles, sanitaires et Ã©conomiques.
Cela peut Ãªtre rÃ©alisÃ© d'abord en crÃ©ant un bloc de donnÃ©es contenant les scores Jaccard pour chaque tweet pour chaque catÃ©gorie, puis en attribuant un tweet Ã  une catÃ©gorie en fonction du score le plus Ã©levÃ© et enfin en regroupant les tweets par nom d'utilisateur et somme des tweets.

```python
def jaccard_similarity(query, document):
    intersection = set(query).intersection(set(document))
    union = set(query).union(set(document))
    return len(intersection)/len(union)
def get_scores(group,tweets):
    scores = []
    for tweet in tweets:
        s = jaccard_similarity(group, tweet)
        scores.append(s)
    return scores
e_scores = get_scores(economy, df.Tweet.to_list())
s_scores = get_scores(social, df.Tweet.to_list())
c_scores = get_scores(culture, df.Tweet.to_list())
h_scores = get_scores(health, df.Tweet.to_list())
```

L'indice de Jaccard, Ã©galement connu sous le nom d'intersection sur l'union et le coefficient de similitude de Jaccard (Ã  l'origine donnÃ© le nom franÃ§ais coefficient de communautÃ© par Paul Jaccard), est une statistique utilisÃ©e pour mesurer la similitude et la diversitÃ© des ensembles d'Ã©chantillons. Le coefficient de Jaccard mesure la similitude entre les ensembles d'Ã©chantillons finis et est dÃ©fini comme la taille de l'intersection divisÃ©e par la taille de l'union des ensembles d'Ã©chantillons.

```python
# create a jaccard scored df.
data  = {'names':df.TweetId.to_list(),       'economic_score':e_scores,
         'social_score': s_scores, 'culture_score':c_scores, 'health_scores':h_scores}
scores_df = pd.DataFrame(data)
#assign classes based on highest score
def get_classes(l1, l2, l3, l4):
    econ = []
    socio = []
    cul = []
    heal = []
    for i, j, k, l in zip(l1, l2, l3, l4):
        m = max(i, j, k, l)
        if m == i:
            econ.append(1)
        else:
            econ.append(0)
        if m == j:
            socio.append(1)
        else:
            socio.append(0)        
        if m == k:
            cul.append(1)
        else:
            cul.append(0)  
        if m == l:
            heal.append(1)
        else:
            heal.append(0)   
            
    return econ, socio, cul, heal
l1 = scores_df.economic_score.to_list()
l2 = scores_df.social_score.to_list()
l3 = scores_df.culture_score.to_list()
l4 = scores_df.health_scores.to_list()
econ, socio, cul, heal = get_classes(l1, l2, l3, l4)
data = {'name': scores_df.names.to_list(), 'economic':econ, 'social':socio, 'culture':cul, 'health': heal}
class_df = pd.DataFrame(data)
#grouping the tweets by username
new_groups_df = class_df.groupby(['name']).sum()
#add a new totals column
new_groups_df['total'] = new_groups_df['health'] + new_groups_df['culture'] + new_groups_df['social'] +  new_groups_df['economic']
#add a new totals row
new_groups_df.loc["Total"] = new_groups_df.sum()
```
